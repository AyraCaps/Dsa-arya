>Case study - problem



A company started to invest in digital marketing as a new way of their product promotions. For that they collected data and decided to carry out a study on it.

*   The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.
*   The company needs to check whether there is any dependency between the features “Region” and “Manager”.






```python
#importing the necessary libraries
import numpy as np
import pandas as pd

import scipy.stats as stats
from scipy.stats import ttest_rel
from scipy.stats import chi2_contingency
```

```python
#Loading the data
sales_data=pd.read_csv('/content/Sales_add.csv')
sales_data.head()
```

***Performing some descriptive analysis and Null Value check.***

```python
sales_data.isna().sum()
```

**Conducting Case study- hypothesis testing**

 **Case 1 : Whether there is any increase in sales after stepping into digital marketing.**




Here the sample size is less than 30,so we prefer t test.  To clarify whether there is any increase in sales after stepping into digital marketing we can use paired t test (because paired t test is more convinient when it comes to  before and after cases).

Define the Null Hypothesis (H0) and Alternate Hypothesis(Ha) and  set the level of Significance as alpha = 0.05.

*  H0 : there is no increase in sales after stepping into digital marketing.

*  Ha : there is an increase in sales after stepping into digital marketing.







```python
# sales before and after of digital advertisment is stored into new varrianles sales_old and sales_new respectively
sales_old=sales_data['Sales_before_digital_add(in $)']
sales_new=sales_data['Sales_After_digital_add(in $)']

#performing paired t- test
t_stat,p_value = ttest_rel(sales_old,sales_new)
```

```python
print(' p-value =', p_value , '\n test statistic value=' ,t_stat)
```

*For decision making we need to compare p value with the level of significance value and if p is less than significance level, null hypothesis will be rejected, otherwise accepted*.

```python
#deciding reject or accept the Null hypothesis based on p-value
if p_value<0.05:
  print('Reject Null Hypothesis')
else:
  print('Accept Null Hypothesis')
```


We can say that there is a significant increase in sales after doing Digital advertisements.

 ***Case 2 : Checking whether there is any dependency between the features “Region” and “Manager”***

Here we conducting the Chi-square test for independence of features( region and manger).

Define the Null Hypothesis (H0) and Alternate Hypothesis(Ha) and  set the level of Significance as alpha = 0.05.

*  H0 : there is no dependancy between the features 'Region and Manager

*  Ha : there is  dependancy between the features 'Region and Manager







```python
#Creating a crosstab table of Region and Manager
data_crosstab = pd.crosstab(sales_data["Region"],sales_data["Manager"])
data_crosstab
```

```python
# Performing Chi-square test for independence of features
from scipy.stats import chi2_contingency
chi_stat,p_value,dof,expected = chi2_contingency(data_crosstab)
```

```python
print(' p-value =', p_value , '\n test statistic value=' ,t_stat, '\n degrees of freedom=',dof)
```

 For decision making we need to compare p value with the level of significance value and if p is less than significance level, null hypothesis will be rejected, otherwise accepted

```python
#deciding reject or accept the Null hypothesis based on p-value
if p_value<0.05:
  print('Reject Null Hypothesis')
else:
  print('Accept Null Hypothesis')
```


Hence, we can conclude that there is no dependancy between the features 'Region' and 'Manager'
