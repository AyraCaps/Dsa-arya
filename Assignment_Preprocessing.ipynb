***Handling Missing values***


```python
# To check  missing values
data.isna().sum()
```

there are missing values in Age, Cabin, Embarked columns.

```python
num_columns = data.select_dtypes(include=['float64','int64'])
num_columns.hist(figsize=(20,15));
```

```python
#Calculate the skew for 'AGE' column
data['Age'].skew()

#if the skewness value < -1 or > +1 -- Highly skewed distribution
#if the skewness value is between -1 & -1/2 or between +1/2 & +1 -- Moderately␣ skewed distribution
#if the skewness value between -1/2 & +1/2 -- Approximately symmetric distribution
```

 Here the skewness  value is between -1/2 & +1/2  hence Approximately symmetric.
 hist plot of age shows thai it is slightly skewed.

```python
# To replace missing values
data.Age=data['Age'].fillna(data['Age'].median()) # replace the missing values with median.
data.Cabin=data['Cabin'].fillna((data['Cabin'].mode()[0])) # we can replace the missing values with mode, since it is a categorical data.
data.Embarked=data['Embarked'].fillna((data['Embarked'].mode()[0])) #we can replace the missing values with mode, since it is a categorical data.
```

 ***Check and handle Outliers in at least 3 columns in the dataset***

```python
# checking outliers in 'Age','SibSp' and 'Fare' columns in the dataset

for i in data[['Age','SibSp','Fare']]:
  plt.figure()
  plt.boxplot(data[i])  # BOXPLOT
  plt.title(i);
```

outliers present in 'Age' ,'SibSp', 'Fare' column

```python
for i in data[['Age','SibSp','Fare']]:
  skew=data[i].skew()
  print(skew)
```

```python
# outlier handling using IQR method
for i in data[['Age','SibSp','Fare']]:
  print('  ', i)
  print('  ----')
  print('Outlier handling of', i)
  #Calculating the interquartile range
  Q1 = np.percentile(data[i],25,method='midpoint')
  Q2 = np.percentile(data[i],50,method='midpoint')
  Q3 = np.percentile(data[i],75,method='midpoint')
  IQR =Q3 - Q1

  print("Q1 is: ",Q1)
  print("Q2 is: ",Q2)
  print("Q3 is: ",Q3)
  print('IQR =',IQR)

  #Finding the lower limit and upper limit
  low_lim = Q1 - 1.5*IQR
  up_lim = Q3 +1.5*IQR
  print('lower limit = ',low_lim )
  print('upper limit  ',up_lim )

  #Finding outliers
  outliers =[]

  for x in data[i]:
    if((x < low_lim) or (x > up_lim)):
      outliers.append(x)


  print('outliers : ',outliers)
 # upper values and lower values are replaced by upper limit and lower limit (respectively)by using clip function
  data[i] = data[i].clip(lower=low_lim ,upper=up_lim)



  print('                        #--------------------------------------------------------------#')



```

```python
#checking outliers after handing it by IQR method
print('outliers after IQR')
for i in data[['Age','SibSp','Fare']]:
  outliers_clip =[]

  for x in data[i]:
    if((x < low_lim) or (x > up_lim)):
      outliers.append(x)

  print(i, ':outliers= '  ,outliers_clip)
```
