```python
import pandas as pd
```

```python
# Load the dataset
data=pd.read_csv('/content/StudentsPerformance.csv')
data.head()   # Display the first few rows of the dataset
```

```python
data.info()
```

***1.Males and females participated in the test.***

```python
# Number of males and females participated in the test
data['gender'].value_counts()
```



*   Number of males participated in the test= 518
*   Number of females participated in the test= 482
  



***2.The students' parental level of education***

```python
# display the parental level of education of the students
data['parental level of education'].unique()
```

```python
# average marks of students based on their parental level of eduction
parent_education=data.groupby('parental level of education').mean()
parent_education.sort_values(['math score','reading score','writing score'],ascending=False)
```



*   Parental level of education: from high school to a master’s degree


*   Students  with a master’s degree as the parental level of education have highest average scores in maths , readind and writing.

*   Students  with  high school as the parental level of education have lowest average scores in maths , readind and writing.






***3. Average score on each subject w.r.t gender and Test preparation course***

```python
#average score based on gender
data.groupby('gender')[['math score','reading score','writing score']].mean()
```


*   Female students have higher average
reading and writing scores than their male counterparts. Where male students have higher average math scores

```python
#average score based on test preparation course
data.groupby('test preparation course').mean()
```

* Those who completed the Test preparation course scores higher average score for math, reading and writing than the ones who did not take any test preparation course.

***4. Scoring variation based on gender and Test preparation course***

Based on the standard deviation ,which is a measure of the variability or dispersion of scores within a dataset we can find the relative performance of students based on their gender and test preparation course.

```python
data.groupby('gender').std()
```



*    Standard deviation of scores of the male students is low, which indicates less variability. So they are more consistent in their performance than female students.



```python
data.groupby('test preparation course').std()
```



*   Standard deviation of scores of the  students with those who completed the Test preparation course is low, which indicates less variability. So they are more consistent in their performance than the ones who did not take any test preparation course.





***5.The top 25% of students based on their Math score***





```python
#Select the students with Math score in the top 25%
maths_top=data['math score'].quantile(0.75)
Top25_Percent=data[data['math score']>=maths_top]
# students getting bonus points
bonus_point_students=Top25_Percent.sort_values('math score',ascending=False)
bonus_point_students
```
